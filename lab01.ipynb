```python
people=['sindhu','sathvika','vyshu','lakshmi','prash']
age=[21,19,22,34,24]
weight=[45,47,43,50,45]
height=[160,155,165,145,150]
```

```python
import matplotlib.pyplot as plt
plt.scatter(weight,height)
plt.title("Relationship between weight and height of students")
plt.xlabel("weight in kgs")
plt.ylabel("height of students in kgs")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(age,height)
plt.title("Relationship between age and height of students")
plt.xlabel("age of students")
plt.ylabel("height of students in kgs")
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.title("People's weight in kgs",fontsize=16,fontstyle='italic')
#main plot function 'bar'
plt.bar(x=people,height=weight,width=0.6,color='orange',edgecolor='k',alpha=0.6)
plt.xlabel("People",fontsize=15)
plt.xticks(fontsize=14,rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("weight(in kgs)",fontsize=15)
plt.show()
```
